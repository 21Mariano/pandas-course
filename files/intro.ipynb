```python
import pandas as pd
```

```python
# Create a sample DataFrame
"""
- DataFrame tiene 2 parametros principales:
  1. data: se ponen entre corchetes cada fila de la tabla
  2. columns: se coloca columns=[] y se ponen entre corchetes los nombres de las columnas
  3. index: se coloca index=[] y se ponen entre corchetes los nombres de las filas (opcional)
"""

dataRows = [["mariano", "reyes", 18], ["ana", "gutierrez", 20], ["john", "doe", 20], ["clara", "rodriguez", 20], ["mariano", "reyes", 20]]

df = pd.DataFrame(dataRows, columns=["name", "surname", "age"])

# Get the last 3 items of the DataFrame
finalItem = df.tail(2)

# Get the first 2 items of the DataFrame
firstItem = df.head(2)

# Get unique rows from the 'name' column
dataRowA = df["name"].unique()

# Get the shape of the DataFrame
dataShape = df.shape

# get the size of the DataFrame (dot product of rows and columns)
dataSize = df.size

print(firstItem)
print("- - - - - - - - - - - - - - - - -")
print(finalItem)
print("- - - - - - - - - - - - - - - - -")
print(dataRowA)
print("- - - - - - - - - - - - - - - - -")
print(dataShape)
print("- - - - - - - - - - - - - - - - -")
print(dataSize)
```
