```python
import pandas as pd
```

```python
clients = pd.read_csv("../data/clients.csv")
orders = pd.read_csv("../data/orders.csv")

clients.rename(columns={
  "cliente_id": "client_id",
  "nombre": "name",
  "pais": "country"
}, inplace=True)

orders.rename(columns={
  "pedido_id": "order_id",
  "cliente_id": "client_id",
  "producto": "product",
  "monto": "amount"
}, inplace=True)
```

```python
pd.merge(clients, orders, on="client_id", how="inner")
```

```python
pd.merge(clients, orders, on="client_id", how="right")
```

```python
# Visualizar clientes que no realizaron pedidos
no_orders_merge = pd.merge(clients, orders, on="client_id", how="left")
no_orders_merge.loc[no_orders_merge["order_id"].isna(), ["client_id", "name", "country"]]

# no_orders_isin = clients.loc[~clients["client_id"].isin(orders["client_id"]), ["client_id", "name", "country"]]
# no_orders_isin
```

```python
# Clientes no registrados en clients.csv pero que realizaron pedidos
clients_merged = pd.merge(clients, orders, on="client_id", how="right", indicator=True)
clients_not_registered = clients_merged.loc[clients_merged["name"].isna(), ["client_id"]]

clients_not_registered
```

```python
# Merge both data frames by "client_id"
merged_df = pd.merge(clients, orders, on="client_id")

# Calculate the sum of amounts by country
amount_by_country = merged_df.groupby("country")["amount"].sum().reset_index(name="total_amount")

amount_by_country
```
