```python
import pandas as pd
```

```python
results = pd.read_csv("../data/results_not_cleaned.csv")
```

```python
columns_to_keep = ["Games", "Event", "Pos", "As", "athlete_id", "NOC", "Discipline", "Medal"]
results = results[columns_to_keep]
```

```python
results.loc[results["Medal"].isna(), "Medal"] = None
```

```python
results.rename(columns={
  col: col.lower() for col in columns_to_keep
}, inplace=True)
```

```python
results.drop_duplicates(inplace=True)
results.reset_index(drop=True, inplace=True)
```

```python
results = results[results["pos"].notna()]
results = results[results["noc"].notna()]
results = results[results["discipline"].notna()]
```

```python
results.reset_index(drop=True, inplace=True)
```

```python
results["pos"] = results["pos"].str.replace("=", "").str.replace(r".0", "", regex=True)
results = results[
    results["pos"].isin(["DNF", "DNS"]) | 
    results["pos"].str.match(r'^\d+$')
]
```

```python
results.reset_index(drop=True, inplace=True)
```

```python
results.to_parquet("../data/results_cleaned.parquet")
```
