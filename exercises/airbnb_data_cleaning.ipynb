```python
import pandas as pd
```

```python
data = pd.read_csv('../data/airbnb.csv')
```

```python
data.drop(columns=["reviews per month", "last review", "house_rules", "license"], inplace=True)
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.dropna(inplace=True)
```

```python
data.drop(columns=["neighbourhood group", "host name", "country code", "availability 365", "Construction year", "calculated host listings count", "host_identity_verified"], inplace=True)
```

```python
columns = data.columns
new_columns = []
for column in columns:
  new_columns.append(column.replace(" ", "_").lower())

data.columns = new_columns
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
data["price"] = data["price"].str.replace("$", "").str.replace(",", ".").astype(float)
data["service_fee"] = data["service_fee"].str.replace("$", "").str.replace(",", ".").astype(float)
```

```python
data["instant_bookable"] = data["instant_bookable"].apply(lambda x: 1 if x == True else 0)
```

```python
data.to_csv('../data/airbnb_cleaned.csv', index=False)
```
