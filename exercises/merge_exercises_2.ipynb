```python
import pandas as pd
```

```python
products = pd.read_csv("../data/products.csv")
sales = pd.read_csv("../data/sales.csv")

sales.rename(columns={
  "venta_id": "sale_id",
  "producto": "product",
  "monto": "amount",
  "fecha": "date"
}, inplace=True)
sales["date"] = pd.to_datetime(sales["date"])

products.rename(columns={
  "producto_id": "product_id",
  "producto": "product",
  "categoria": "category"
}, inplace=True)
```

```python
pd.merge(products, sales, on="product", how="inner")
```
